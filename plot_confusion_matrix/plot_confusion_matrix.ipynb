## Usage

#### Load the necessary libraries

```python
import pandas as pd
from plot_confusion_matrix import plot_confusion_matrix
from sklearn.metrics import confusion_matrix
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
# Not needed in your notebook. This is just for demo
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
```

```python
#Load iris dataset
iris = load_iris()
X = iris.data
y = iris.target

n_features = X.shape[1]
```

```python
# Fit logistic
lr = LogisticRegression()
lr.fit(X,y)
```

```python
pred = lr.predict(X)
```

### Creating the Confusion Matrix

```python
cnf_matrix = confusion_matrix(y, pred)
plt.figure(figsize=(5,4))

plot_confusion_matrix(cnf_matrix, classes=[0,1,2],title='Confusion matrix')
```

```python
#Change color
plot_confusion_matrix(cnf_matrix, classes=[0,1,2],title='Confusion matrix',cmap=plt.cm.Reds)
```

```python
#Change color
plot_confusion_matrix(cnf_matrix, classes=[0,1,2],title='Confusion matrix',cmap=plt.cm.Greens)
```
